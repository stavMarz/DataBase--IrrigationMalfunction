
 **Imports**

```python
import pandas as pd
import numpy
from PIL import Image
import os
from os import listdir
import sys
```

**Fix Longitude and Latitude**

```python

csvdir = os.getcwd()
for file in os.listdir(csvdir):
    if(file.endswith('.csv')):
        path = os.path.join(csvdir,file)
        print(path)
        new = pd.read_csv(path)
        new['Latitude'] = new['Latitude']+6
        new['Longitude'] = new['Longitude']+3

        name = ('norm'+file)
        savepath = os.path.join(csvdir,name)
        new.to_csv(savepath)
```
